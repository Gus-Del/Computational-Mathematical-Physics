\\( f'(x)=\lim_{h\rightarrow 0}\frac{f(x+h)-f(x)}{h}\\)

\\( f'(x)=\lim_{h\rightarrow 0}\frac{f(x)-f(x-h)}{h}\\)

\\( f'(x)=\lim_{h\rightarrow 0}\frac{f(x+h)-f(x-h)}{2h}\\)

All above are equivalent in case of a continuous function.

First-order derivatives, numerical implementation

Forward differences
\\( f'(x_n)\approx\frac{f(x_{n+1})-f(x_n)}{x_{n+1}-x_n}\\)

Backward differences
\\( f'(x)\approx\frac{f(x_n)-f(x_{n-1})}{x_n-x_{n-1}}\\)

Central differences
\\( f'(x)\approx\frac{f(x_{n+1})-f(x_{n-1})}{x_{n+1}-x_{n-1}}\\)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#We start by defining our function. 
def f(x):
    return x
```

```python
x_list = np.linspace(-10,10, 100) #Generates 100 evenly spaced values between -10 to 10. 
y_list = f(x_list)

plt.plot(x_list, y_list)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
x_list = np.linspace(-10,10, 100)
d_list = np.ones_like(x_list)
#We can't use d_list = 1 b/c this would define a scalar when we need an array. 
#np.ones_like(x_list) creates an array of ones with the same shape as x_list, since the derivative is 1 for all x. 

plt.plot(x_list, d_list)
plt.xlabel('x')
plt.ylabel("f'(x)")
```

### Example function: \\( f(x)=\cos(x)x-\frac{1}{100}x^2 \\)

```python
def f(x):
    return np.cos(x)*x - 1/100 * x**2
```

```python
x_list = np.linspace(-10,10, 100) #Generates 100 evenly spaced values between -10 to 10. 
y_list = f(x_list)

plt.plot(x_list, y_list)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
x_list = np.linspace(-10,10, 100)
d_list = np.sin(x_list)*(-1)*x_list + np.cos(x_list) - 2/100*x_list

plt.plot(x_list, d_list)
plt.xlabel('x')
plt.ylabel("f'(x)")
```

```python
#Now, adding the numerical derivatives. 

h = 0.1
d_list_forward = ( f(x_list+h) - f(x_list)) / h
d_list_backward = ( f(x_list) - f(x_list-h)) / h
d_list_central = ( f(x_list+h) - f(x_list-h)) / (2*h)

plt.plot(x_list, d_list_forward)
plt.plot(x_list, d_list_backward)
plt.plot(x_list, d_list_central)
```

Notice the small differences between the plots. 

```python
#Now we will analyze this is in more detail by analyzing the error. 
plt.xlabel('x')
plt.ylabel("Error of f'(x)")
# plt.plot(x_list, d_list)

plt.plot(x_list, d_list - d_list_forward, 'red')
plt.plot(x_list, d_list - d_list_backward, 'blue')
plt.plot(x_list, d_list - d_list_central, 'green')
```

Notice above how the Central Differences Method(green) is more accurate than the the Forward and Backward methods. 

# 2nd Derivatives

We derive \\(f'(x)\\) another time

\\( f''(x)=\lim_{h\rightarrow 0}\frac{f'(x+h)-f'(x)}{h}\\)

\\( f''(x)=\lim_{h\rightarrow 0}\frac{f'(x)-f'(x-h)}{h}\\)

\\( f''(x)=\lim_{h\rightarrow 0}\frac{f'(x+h)-f'(x-h)}{2h}\\)

```python
x_list = np.linspace(-10,10, 200)
#d_list = np.sin(x_list)*(-1)*x_list + np.cos(x_list) - 2/100*x_list
d2_list = -x_list*np.cos(x_list)-2*np.sin(x_list) - 2/100        #2nd derivative of function.

h = 0.1
d2_list_forward = ( f(x_list+2*h) - 2*f(x_list+h) + f(x_list)) / h**2
d2_list_backward = ( f(x_list) - 2*f(x_list-h) + f(x_list- 2*h)) / h**2
d2_list_central = ( f(x_list+h) - 2*f(x_list) + f(x_list-h)) / h**2

plt.xlabel('x')
plt.ylabel("f''(x)")
plt.plot(x_list, d2_list)
plt.plot(x_list, d2_list_forward)
plt.plot(x_list, d2_list_backward)
plt.plot(x_list, d2_list_central)
```

```python
plt.xlabel('x')
plt.ylabel("Error f''(x)")

plt.plot(x_list, d2_list - d2_list_forward, 'red')
plt.plot(x_list, d2_list - d2_list_backward, 'blue')
plt.plot(x_list, d2_list - d2_list_central, 'green')
```
