Calculate the velocity \\(v(t)=\dot{x}(t)\\) and the acceleration \\(a(t)=\ddot{x}(t)\\) for the loaded data set that describes a set of points \\((t_i,x_i)\\) for \\(i=0,\dots,1000\\).

Calculate \\(v_i\\) and \\(a_i\\) using any method.


We do not know the function \\(x(t)\\) but only its values for specific points. 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = np.loadtxt("04b_Exercise_velocity_acceleration_data_file.dat")
```

```python
data
```

```python
plt.xlabel('Time t [s]')
plt.ylabel('Coordinate x [m]')
plt.scatter(data[:,0],data[:,1])
```

### Solution

```python
data[:,0]
```

```python
def central(data):
    derive=np.zeros(len(data[:,1]))
    derive[1:-1] = (data[2:,1] - data[:-2,1]) / (data[2:,0] - data[:-2,0])
    #just set to second value or first value use forward method. 
    derive[0] = (data[1,1] - data[0,1]) / (data[1,0] - data[0,0])
    #just set to second-last value or last value use backward method
    derive[-1] = (data[-1,1]-data[-2,1]) / (data[-1,0] - data[-2,0])
    return np.transpose([data[:,0], derive])
```

```python
Centralvelocity = central(data)
Centralacceleration = central(central(data))
```

```python
plt.xlabel('Time t [s]')
plt.ylabel('Velocity v [m/s]')
plt.scatter(Centralvelocity[:,0], Centralvelocity[:,1])
plt.show()

plt.xlabel('Time t [s]')
plt.ylabel('Acceleration a [m/(s^2)]')
plt.scatter(Centralacceleration[:,0],Centralacceleration[:,1])
plt.show()
```
