```python
# Example 1: Simple power rule
import sympy
x = sympy.Symbol('x')                 #defining 'x' and 'y' as symbolic variables you're working with.
y = sympy.Symbol('y')

integral1 = sympy.integrate(y*x**2, x)  #This is where the integration happens.
                                      #sympy.integrate is the function used to perform symbolic integration.
                                      #The first argument (x**2) is the expression to be integrated (x raised to the power of 2).
                                      #The second argument (x) specifies the variable with respect to which we're integrating.

print(f"Integral of y*x^2: {integral1}") #Finally, this line prints the result of the integration.
                                       #integral1 will store the calculated indefinite integral of x**2.
                                       #The f-string formatting is used to nicely display the result.
```

```python
# Example 2: Integral of a constant
integral2 = sympy.integrate(5, x)
print(f"Integral of 5: {integral2}")
```

```python
# Example 3: Integral involving trigonometric function
integral3 = sympy.integrate(sympy.sin(x), x)
print(f"Integral of sin(x): {integral3}")


# Example 4: Integral with exponential function
integral4 = sympy.integrate(sympy.exp(x), x)
print(f"Integral of exp(x): {integral4}")

# Example 5:  Integration by parts (more complex)
integral5 = sympy.integrate(x * sympy.cos(x), x)
print(f"Integral of x*cos(x): {integral5}")


# Example 6:  Definite integral with limits
integral6 = sympy.integrate(x**3 , (x, 0, 2))
print(f"Definite integral of x^3 from 0 to 2: {integral6}")


# Example 7:  More complex trigonometric integral 
integral7 = sympy.integrate(sympy.sin(x)**2, x)
print(f"Integral of sin^2(x): {integral7}")

# Example 8: Improper Integral
integral8 = sympy.integrate(1/x**2, (x, 1, sympy.oo))
print(f"Improper integral of 1/x^2 from 1 to infinity: {integral8}")
```

```python
import sympy
import numpy as np
import matplotlib.pyplot as plt

# Define the symbolic variable
x = sympy.Symbol('x')

# Define the function
f = sympy.cos(x) + x**3 + sympy.sin(x)

# Convert the function to a numerical function for plotting
f_numeric = sympy.lambdify(x, f, 'numpy')

# Generate x values
x_vals = np.linspace(-5, 5, 400)  # Increased the range for better visualization

# Calculate the corresponding y values
y_vals = f_numeric(x_vals)

# Plot the function
plt.plot(x_vals, y_vals, label='cos(x) + x^3 + sin(x)')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of cos(x) + x^3 + sin(x)')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import sympy
import numpy as np
import matplotlib.pyplot as plt

# Define the symbolic variable
x = sympy.Symbol('x')

# Define the function
f = sympy.cos(x) + x**3 + sympy.sin(x)

# Convert the function to a numerical function
f_numeric = sympy.lambdify(x, f, 'numpy')

# Generate x values from -15 to 15
x_vals = np.linspace(-15, 15, 200)

# Calculate the corresponding y values
y_vals = f_numeric(x_vals)

# Plot the function
plt.plot(x_vals, y_vals, label='f(x) = cos(x) + x^3 + sin(x)')

# Fill the area under the curve from -10 to 10.
plt.fill_between(x_vals, y_vals,0, where=(x_vals >= -10) & (x_vals <= 10), alpha=0.5, color='skyblue', label='Area from -10 to 10')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) with Filled Area')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
#plot the integral of the above function f = sympy.cos(x) + x**3 + sympy.sin(x)

import sympy
import numpy as np
import matplotlib.pyplot as plt

x = sympy.Symbol('x')
f = sympy.cos(x) + x**3 + sympy.sin(x)

# Calculate and print the definite integral from -pi to pi
definite_integral = sympy.integrate(f, (x, -sympy.pi, sympy.pi))
print(f"Definite Integral from -pi to pi: {definite_integral}")

# Plot the integral from -5 to x
area_function = sympy.integrate(f, (x, -5, x))
area_f = sympy.lambdify(x, area_function, 'numpy')
x_vals = np.linspace(-10, 10, 200)
y_integral = area_f(x_vals)

plt.plot(x_vals, y_integral, label='Integral of f(x) from -5 to x')
plt.xlabel('x')
plt.ylabel('Integral')
plt.title('Plot of Integral')
plt.legend()
plt.grid(True)
plt.show()
```
